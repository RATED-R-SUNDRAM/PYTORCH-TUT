### THE BROADER IDEA OF DATASET AND DATALOADER CLASS IS SIMPLE AS SUMMARIZED ABOVE ALSO

* THE DATASET IS A ABSTRACT CLASS WHICH HAS THE 3 FUNCTIONS init for laoding data , len for length and getitem for getting item at an index, these functions are important as dataloader interacts with these functions

* The dataloader then creates trasformations like shuffling, batch creations etc

```python
# CREATING A SAMPLE DATASET

from sklearn.datasets import make_classification
```

```python
x,y= make_classification(
    n_samples=1000,
    n_features=16,
    n_classes=2,
    n_informative=16,
    n_redundant=0,
    random_state=1
)
```

```python
x= torch.tensor(x, dtype=torch.float32)
y= torch.tensor(y,dtype=torch.float32)
```

```python
from torch.utils.data import Dataset , DataLoader
```

```python
class customdata(Dataset): # load the abstract Dataset class
  def __init__(self, x,y):
    self.x= x
    self.y= y
  def __len__(self):
    return x.shape[0]
  def __getitem__(self, index):
    return self.x[index], self.y[index]
```

## ONE VERY IMPORTANT THING TO NOTE IS THAT THE DATALOADER USES GETITEM FUNCTION OF DATSET INHERITED CLASS SO IT IS VERY IMPORTANT TO NOTE THAT ANY TRANSFORMATION NEEDED WHILE CREATING BATCHES WITH **TRANSFORMATION SHOULD BE EXECUTED IN GETITEM FUNCTION** OF DATASET INHERITED CLASS

```python
data= customdata(x,y)
```

```python
dataloader = DataLoader(data, batch_size=50, shuffle=True)
```

```python
for batch_feature,batch_label in dataloader:
  print(batch_feature.shape)
  print(batch_label.shape)
```
